```python
# top 20 words before removing stopwords bar chart
from sklearn.feature_extraction.text import CountVectorizer

def get_top_n_words(corpus, n=None):
    vec = CountVectorizer().fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0)
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq = sorted(words_freq, key = lambda x:x[1], reverse=True)
    return words_freq[:n]

common_words = get_top_n_words(dataset["Description"], 20)
dataframe1 = pd.DataFrame(common_words, columns = ['Description', 'count'])
dataframe1.groupby('Description').sum()['count'].sort_values(ascending=False).plot(kind='bar',
                                                                                ylabel='Count',
                                                                                title='Top 20 words in articles before removing stopwords');
```

```python
def get_top_n_bigram(corpus, n=None):
    vec = CountVectorizer(ngram_range=(2, 2), stop_words='english').fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0)
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq = sorted(words_freq, key =lambda x:x[1], reverse=True)
    return words_freq[:n]

common_words = get_top_n_bigram(dataset["Description"], 20)
dataframe2 = pd.DataFrame(common_words, columns = ['bigram', 'count'])
dataframe2.groupby('bigram').sum()['count'].sort_values(ascending=False).plot(kind='bar',
                                                                              ylabel='Count',
                                                                              title='Top 20 bigrams in articles after removing stop words and lemmatization');

```

```python
# Install XGBoost 
!pip install xgboost

# Import necessary libraries
from xgboost import XGBClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report

# Ensure that necessary libraries for resampling (SMOTE) are imported
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split

# Assuming `X_train` and `y_train` are your original training data
# Apply SMOTE to resample the imbalanced dataset
smote = SMOTE(random_state=42)
x_train_resampled, y_train_resampled = smote.fit_resample(x_train, y_train)

# Step 1: Encode string labels into numerical values using LabelEncoder
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train_resampled)  # Encode SMOTE-resampled training labels
y_test_encoded = label_encoder.transform(y_test)  # Encode test labels using the same encoder

# Step 2: Initialize XGBoost Classifier with reasonable hyperparameters
xgb_model = XGBClassifier(
    n_estimators=100,  # Number of trees
    max_depth=5,  # Maximum depth of each tree
    learning_rate=0.1,  # Learning rate (eta)
    use_label_encoder=False,  # Turn off label encoder (for XGBClassifier version 1.3.0+)
    eval_metric='mlogloss',  # Metric for multi-class classification
    random_state=42
)

# Step 3: Train the model on the SMOTE-resampled training data (using encoded labels)
xgb_model.fit(x_train_resampled, y_train_encoded)

# Step 4: Make predictions on the original test data
y_pred = xgb_model.predict(x_test)

# Step 5: Decode the predicted numerical values back to the original string labels
y_pred_labels = label_encoder.inverse_transform(y_pred)


```

```python
def find_90_percent_related_words(documents):
  """Finds words that appear in at least 90% of the documents.

  Args:
    documents: A list of lists of words, representing the documents.

  Returns:
    A list of words that appear in at least 90% of the documents.
  """

  from collections import Counter

  # Create a term frequency matrix
  term_frequency_matrix = [Counter(doc) for doc in documents]

  # Calculate document frequency
  document_frequency = Counter()
  for doc_counter in term_frequency_matrix:
    document_frequency.update(set(doc_counter.keys()))

  # Determine 90% threshold
  num_documents = len(documents)
  threshold = int(0.2 * num_documents)

  # Identify 90% related words
  related_words = [word for word, freq in document_frequency.items() if freq >= threshold]

  return related_words
```

```python
# import regular expression and natural language tool kit
import re
import nltk

nltk.download('stopwords')

from nltk.corpus import stopwords 

all_stopwords = stopwords.words('english')
all_stopwords.remove('not')

tokenized_corpus =[]

for i in range(len(news)):
    statement = re.sub('[^a-zA-Z]', ' ', news['text'][i])
    statement = statement.lower()
    statement = statement.split()
    statement = [word for word in statement if word not in all_stopwords]
   
    tokenized_corpus.append(statement)
```

```python
related = find_90_percent_related_words(news)
print(related)
```

```python
import gensim
from gensim import corpora, models
# Step 2: Create Dictionary and Corpus (BoW format)
id2word = corpora.Dictionary(tokenized_corpus)  # Create dictionary of unique words
corpus_bow = [id2word.doc2bow(text) for text in tokenized_corpus]  # Create corpus (BoW format)
```

```python
# Build LDA model
lda_model_words = gensim.models.LdaMulticore(corpus=corpus_bow,
                                       id2word=id2word,
                                       num_topics=7, 
                                       random_state=100,
                                       chunksize=100,
                                       passes=10,
                                       per_word_topics=True)
```

```python
import matplotlib.pyplot as plt
import math
from matplotlib import gridspec
k =10
def plot_top_words(lda=lda_model_words , nb_topics=7, nb_words=10):
    top_words = [[word for word,_ in lda.show_topic(topic_id, topn=50)] for topic_id in range(lda.num_topics)]
    top_betas = [[beta for _,beta in lda.show_topic(topic_id, topn=50)] for topic_id in range(lda.num_topics)]

    gs  = gridspec.GridSpec(round(math.sqrt(k))+1,round(math.sqrt(k))+1)
    gs.update(wspace=0.5, hspace=0.5)
    plt.figure(figsize=(20,15))
    for i in range(nb_topics):
        ax = plt.subplot(gs[i])
        plt.barh(range(nb_words), top_betas[i][:nb_words], align='center',color='blue', ecolor='black')
        ax.invert_yaxis()
        ax.set_yticks(range(nb_words))
        ax.set_yticklabels(top_words[i][:nb_words])
        plt.title("News_Topic "+str(i))
        
  
plot_top_words()
```
