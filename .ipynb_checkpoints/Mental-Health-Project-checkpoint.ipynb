### Data Cleaning

```python
dataset.statement
```

```python
# import regular expression and natural language tool kit
import re
import nltk

nltk.download('stopwords')

from nltk.corpus import stopwords 
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

all_stopwords = stopwords.words('english')
all_stopwords.remove('not')
```

```python
corpus = []

for i in range(len(dataset)):
    statement = re.sub('[^a-zA-Z]', ' ', dataset['statement'][i])
    statement = statement.lower()
    statement = statement.split()
    statement = [ps.stem(word) for word in statement if not word in set(all_stopwords)]
    statement = ' '.join(statement)
    corpus.append(statement)
```

### Dividing dataset into training and test set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 0)
```

### Handle imbalanced data set

```python
#!pip install imbalanced-learn

# Import the necessary libraries
from imblearn.over_sampling import SMOTE
from collections import Counter
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt


print("Before SMOTE class distribution:", Counter(y_train))

# Initialize SMOTE
smote = SMOTE(random_state=42)

# Apply SMOTE to the training data (x_train and y_train)
x_train_resampled, y_train_resampled = smote.fit_resample(x_train, y_train)

# Count the number of instances in each class after SMOTE
class_counts = Counter(y_train_resampled)

# Plot the class distribution
labels, values = zip(*class_counts.items())  # Unzip the labels and counts

plt.bar(labels, values, width=0.5)

plt.xlabel('Class Labels')
plt.ylabel('Number of Instances')
plt.title('Class Distribution After SMOTE')
# Rotate x-axis labels for better readability
plt.xticks(labels, rotation=45, ha='right', fontsize=10)

plt.show()

# Print the class distribution after SMOTE
print("After SMOTE class distribution:", Counter(y_train_resampled))
```

### Model Fitting (Naive Bayes)

```python
# Part 1: Model Fitting (MultinomialNB with Hyperparameter Tuning)

from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.preprocessing import LabelEncoder
import joblib

# Step 1: Encode string labels into numerical values
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train_resampled)  # Encode SMOTE-resampled training labels

# Step 2: Define the parameter grid for hyperparameter tuning
param_grid = {'alpha': [0.1, 0.5, 1.0]}  # 'alpha' is the smoothing parameter for Naive Bayes

# Step 3: Initialize StratifiedKFold with random_state for reproducibility
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Step 4: Initialize GridSearchCV for hyperparameter tuning with MultinomialNB
grid_search = GridSearchCV(MultinomialNB(), param_grid, cv=cv, scoring='accuracy', verbose=1, n_jobs=-1)

# Step 5: Fit the MultinomialNB model on the resampled training data
grid_search.fit(x_train_resampled, y_train_encoded)

# Step 6: Get the best model after hyperparameter tuning
best_classifier = grid_search.best_estimator_

# Step 7: Exporting the best MultinomialNB Classifier to later use in prediction
joblib.dump(best_classifier, 'c2_Classifier_Sentiment_Model')


```

### Model Performance (Naive Bayes)

```python
# Part 2: Model Evaluation

from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, classification_report

# Step 1: Encode test labels if they are not encoded already
y_test_encoded = label_encoder.transform(y_test)  # Encode test labels if y_test is in string format

# Step 2: Make predictions on the test data
y_pred = best_classifier.predict(x_test)  # Use the best classifier from the previous step

# Step 3: Decode both y_pred and y_test into string labels for consistency
y_test_decoded = label_encoder.inverse_transform(y_test_encoded)  # Convert y_test back to strings
y_pred_decoded = label_encoder.inverse_transform(y_pred)  # Convert y_pred back to strings

# Step 4: Calculate the confusion matrix and performance metrics
cm = confusion_matrix(y_test_decoded, y_pred_decoded)
print("Confusion Matrix:")
print(cm)

# Accuracy Score
accuracy = accuracy_score(y_test_decoded, y_pred_decoded)
print(f"Accuracy Score: {accuracy:.4f}")

# Precision, Recall, F1-Score
precision = precision_score(y_test_decoded, y_pred_decoded, average='weighted')
recall = recall_score(y_test_decoded, y_pred_decoded, average='weighted')
f1 = f1_score(y_test_decoded, y_pred_decoded, average='weighted')

print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")

# Classification Report
print("\nClassification Report:")
print(classification_report(y_test_decoded, y_pred_decoded))
```

# PART THREE: SUMMARIZTION

```python
from nltk.cluster.util import cosine_distance
import networkx as nx
from nltk.tokenize import sent_tokenize
```

```python
def read_input(text):
    sentences = []
    sentences = sent_tokenize(text)
    for sentence in sentences:
        sentence.replace("[^a-zA-Z0-9]"," ")
    return sentences
```

```python
def sentence_similarity(sent1,sent2,stopwords = None):
    if stopwords is None:
        stopwords = []
    sent1 = [w.lower() for w in sent1]
    sent2 = [w.lower() for w in sent2]
    all_words = list(set(sent1+sent2))
    vector1 = [0]*len(all_words)
    vector2 = [0]*len(all_words)
    #build vector for 1st sentence
    for w in sent1:
        if not w in stopwords:
            vector1[all_words.index(w)]+=1
    for w in sent2:
        if not w in stopwords:
            vector2[all_words.index(w)]+=1
    return 1-cosine_distance(vector1,vector2)
```

```python
def build_similarity_matrix(sentences,stop_words):
    #create an empty similarity matrix
    similarity_matrix = np.zeros((len(sentences),len(sentences)))
    
    for idx1 in range(len(sentences)):
        for idx2 in range(len(sentences)):
            if idx1!=idx2:
                similarity_matrix[idx1][idx2] = sentence_similarity(sentences[idx1],sentences[idx2],stop_words)
                
    return similarity_matrix
```

```python
def generate_summary(text,top_n=1):
    nltk.download('stopwords')
    nltk.download('punkt')
    
    stop_words = stopwords.words('english')
    summarize_text = []
    
    # Step1: read text and tokenize
    sentences = read_input(text)
    
    # Steo2: generate similarity matrix across sentences
    sentence_similarity_matrix = build_similarity_matrix(sentences,stop_words)
    
    # Step3: Rank sentences in similarirty matrix
    sentence_similarity_graph = nx.from_numpy_array(sentence_similarity_matrix)
    scores = nx.pagerank(sentence_similarity_graph)
    
    #Step4: sort the rank and place top sentences
    ranked_sentences = sorted(((scores[i],s) for i,s in enumerate(sentences)),reverse=True)
    
    # Step 5: get the top n number of sentences based on rank    
    for i in range(top_n):
        summarize_text.append(ranked_sentences[i][1])
    
    print("Summary: \n", ". ".join(summarize_text))
```

PART4 :TOPIC MODELING

```python
from sklearn import decomposition
lda = LatentDirichletAllocation(n_components = 7, random_state=0)
lda.fit(x_train)
```

```python
for idx,topic in enumerate(lda.components_):
    print(f"Topic {idx}: ")
    print([cv.get_feature_names_out()[i] for i in topic.argsort()[:-10 - 1:-1] ])
```

```python
def display_topics(model, feature_names, no_top_words):
    topic_dict = {}
    for topic_idx, topic in enumerate(model.components_):
        topic_dict[f"Topic {topic_idx+1}"] = [feature_names[i] for i in topic.argsort()[:-no_top_words - 1:-1]]
    return topic_dict
no_top_words = 10
topics = display_topics(lda, cv.get_feature_names_out(), no_top_words)
# Convert a data frame for better visualization 
topics_df = pd.DataFrame(topics)
topics_df
```

```python
# Get topic probabilities for each document
topic_probabilities = lda.transform(x_train)

topic_probabilities_df = pd.DataFrame(topic_probabilities, columns=[f"Topic {i+1}" for i in range(lda.n_components)])
significant_topic = np.argmax(topic_probabilities_df.values, axis=1)
topic_probabilities_df['dominanr_topic'] = significant_topic

topic_probabilities_df
```
